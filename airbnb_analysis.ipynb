2. Price Distribution Analysis


Analytical Step: Price distribution across all entries.


```python
plt.figure(figsize=(8,6))
sns.histplot(df1['price'], bins=50, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_price = df1.groupby('room_type')['price'].mean().reset_index()
sns.barplot(x='room_type', y='price', data=avg_price)
plt.title('Average Price by Room Type')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='number_of_reviews', y='price', data=df1)
plt.title('Price vs Number of Reviews')
plt.show()
```

```python
top_neighbourhoods = df1['neighbourhood'].value_counts().head(10)
sns.barplot(y=top_neighbourhoods.index, x=top_neighbourhoods.values)
plt.title('Top 10 Neighborhoods by Number of Listings')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df1['availability_365'], bins=30, kde=True)
plt.title('Availability Over 365 Days')
plt.xlabel('Days Available')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df1['minimum_nights'], bins=30, kde=True)
plt.title('Minimum Nights for Booking')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df1['reviews_per_month'].dropna(), bins=30, kde=True)
plt.title('Reviews per Month Distribution')
plt.show()
```

```python
avg_price_by_neighbourhood = df1.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(y=avg_price_by_neighbourhood.index, x=avg_price_by_neighbourhood.values)
plt.title('Top 10 Neighborhoods by Average Price')
plt.show()
```

```python
numeric_df = df1.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
df1['last_review'] = pd.to_datetime(df1['last_review'], errors='coerce')
df1['review_month'] = df1['last_review'].dt.month
sns.countplot(x='review_month', data=df1)
plt.title('Number of Reviews by Month')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='number_of_reviews_ltm', y='price', data=df1)
plt.title('Price vs Number of Reviews in Last Month')
plt.show()
```

```python
license_data = df1['license'].notnull().value_counts()
plt.figure(figsize=(8,6))
sns.barplot(x=license_data.index, y=license_data.values)
plt.title('License Availability')
plt.show()
```

```python
avg_availability_by_neighbourhood = df1.groupby('neighbourhood')['availability_365'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(y=avg_availability_by_neighbourhood.index, x=avg_availability_by_neighbourhood.values)
plt.title('Top 10 Neighborhoods by Average Availability')
plt.show()
```
